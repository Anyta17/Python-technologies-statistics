```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("job_listings.csv")
```

```python
import re

technologies = ["Python", "GIT", "SQL", "REST", "API", "docker", "AWS", "linux", "Django", "PostgreSQL", "Artificial Intelligence", "JS", "Machine Learning", "react", "OOP", "Flask", "NoSQL", "networking", "fullstack", "microservice", "MongoDB", "HTML", "CSS", "algorithms", "DRF", "FastAPI", "asyncio", "Graphql"]
tech_counts = {tech: df["Description"].str.count(tech, flags=re.IGNORECASE).sum() for tech in technologies}
```

```python
plt.figure(figsize=(20, 8))
plt.bar(tech_counts.keys(), tech_counts.values())
plt.xticks(rotation=90)
plt.xlabel("Technologies")
plt.ylabel("Mentions")
plt.title("Technology Popularity in Job Descriptions")
plt.show()
```
